# LinearRegression Ml model

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data= pd.read_csv("E:\\ML DATA\canada_per_capita_income.csv")#reading training data to the model
```

```python
data.head()
```

```python
x=data[['year']]
y=data[['per capita income (US$)']]
```

Graph of actual data provides(Testing data)

```python
plt.scatter(x,y,color="red")
plt.xlabel("Year")
plt.ylabel("per capita income (US$)")
plt.show()
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
y1=lr.predict(x)#predicting the per capitaIncome based on  year
```

```python
plt.scatter(x,y,color='blue')
plt.title("Canada per capita IncomeAnalysis",fontsize=15)
plt.xlabel("Year",fontsize=20)
plt.ylabel("PerCapitaIncome US($)",fontsize=15)
plt.plot(x,y1,color='red')
```

```python
p=pd.read_csv("D:\\testingdata.csv")#reading the testing data
```

```python
p.tail(6)
```

```python
price =lr.predict(p)#predicting the result based on experence(testing data)
```

```python
p['perCapitaIncome']=price
```

```python
p.to_csv("perCapitaReportofCanada.csv",index=False)
```
